```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

df_train=pd.read_csv(r'C:\Users\PoYu\jupyter\Example_and_data\Part02\house_train.csv\train.csv')

train_Y=np.log1p(df_train['SalePrice'])
df=df_train.drop(['Id','SalePrice'],axis=1)
df.head()
```

```python
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
        
print(f'{len(num_features)} Numeric Features : {num_features}\n')     
```

```python
df=df[num_features]
df=df.fillna(-1)#填補空值
MMEncoder=MinMaxScaler()#把值變成0~1
df.head()
```

```python
#顯示GrlivArea 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x=df['GrLivArea'],y=train_Y)
plt.show()

#做線性回歸，觀察分數
train_X=MMEncoder.fit_transform(df)
estimator=LinearRegression()
cross_val_score(estimator,train_X,train_Y,cv=5).mean()
```

```python
#將GrLivArea 限制在200~2500以內,調整離群值
df['GrLivArea']=df['GrLivArea'].clip(800,2500)
sns.regplot(x=df['GrLivArea'],y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X=MMEncoder.fit_transform(df)
estimator=LinearRegression()
cross_val_score(estimator,train_X,train_Y,cv=5).mean()
```

```python
#將GrLivArea限制在800到2500以內，捨棄離群值
keep_indexs=(df['GrLivArea']>800&(df['GrLivArea']<2500))
df=df[keep_indexs]
train_Y=train_Y[keep_indexs]
sns.regplot(x=df['GrLivArea'],y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X=MMEncoder.fit_transform(df)
estimator=LinearRegression()
cross_val_score(estimator,train_X,train_Y,cv=5).mean()
```
