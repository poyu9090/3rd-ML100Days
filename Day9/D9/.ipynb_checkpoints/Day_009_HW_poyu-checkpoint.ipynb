```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = r'C:\Users\PoYu\jupyter\Example_and_data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
dtype_df=app_train.dtypes.reset_index()
dtype_df.columns=["Count","Column Type"]
dtype_df=dtype_df.groupby("Column Type").aggregate('count').reset_index()
dtype_df
```

int_features=[]


```python
int_features=[]
float_features=[]
object_features=[]
for dtype, feature in zip(app_train.dtypes,app_train.columns):
    if dtype =='float64':
        float_features.append(feature)
    elif dtype=='int64':
        int_features.append(feature)
    else:
        object_features.append(feature)
print(f'{len(int_features)} Integer Features : {int_features}\n')
print(f'{len(float_features)} Float Features : {float_features}\n')
print(f'{len(object_features)} Object Features : {object_features}')
```

```python
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin([np.dtype('int64'), np.dtype('float64')]))])
print(" Numbers of remain columns " , len(numeric_columns))
```

```python
New_int_features=[]

for col in int_features:
    if len(list(app_train[col].unique())) >2:
            New_int_features.append(col)

print("欄位值總類大於兩種的有%s個,分別是%s" %(len(New_int_features),New_int_features))
```

```python
New_float_features=[]

for col in float_features:
    if len(list(app_train[col].unique())) >2:
            New_float_features.append(col)

print("欄位值總類大於兩種的有%s個,分別是%s" %(len(New_float_features),New_float_features))
```

```python
for col2 in New_int_features:
    fig,axes = plt.subplots()
    app_train[col2].plot(kind='box',ax=axes)
    axes.set_ylabel('values of tip_pct')
    plt.savefig('p1.png')    # 將繪製的圖形儲存為p1.png
```

```python
for col3 in New_float_features:
    fig,axes = plt.subplots()
    app_train[col3].plot(kind='box',ax=axes)
    axes.set_ylabel('values of tip_pct')
    plt.savefig('p1.png')    # 將繪製的圖形儲存為p1.png
    
```
